```python
red_social = [("Juan", ["Maria", "Pedro", "Luis"]),
              ("Maria", ["Juan", "Pedro", "Juan"]), 
              ("Pedro", ["Juan", "Maria"]),
              ("Luis", ["Juan"])]

# Eliminar duplicados y convertir a tuplas
red_social_set = {(persona, tuple(set(amigos))) for persona, amigos in red_social}

# Convertir el conjunto a una tupla
red_social_tuple = tuple(red_social_set)

# Contar los elementos de cada tupla
element_counts = [(persona, len(amigos)) for persona, amigos in red_social_tuple]
print(element_counts)

# Convertir la lista de cuentas a una tupla
red_social_tuple = tuple(element_counts)

# Encontrar la tupla con el mayor número de amigos
tupla_mayor = max(red_social_tuple, key=lambda x: x[1])

print(tupla_mayor)
```
